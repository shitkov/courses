```python
# 02. Last Digit of a Large Fibonacci Number

def get_fibonacci_last_digit_naive(n):
    if n <= 1:
        return n

    previous = 0
    current  = 1

    for _ in range(n - 1):
        previous, current = current, previous + current

    return current % 10

def get_fibonacci_last_digit_fast(n):
    if n <= 1:
        return n

    previous = 0
    current  = 1

    sum_arr = [0, 1]
    
    previous = 0
    current  = 1

    for _ in range(60 - 2):
        previous, current = current, previous + current
        sum_arr.append(current % 10)

    return sum_arr[n % 60]

# test
flag = True
while flag:
    n = random.randint(0, 20)
    res_naive = get_fibonacci_last_digit_naive(n)
    res_test = get_fibonacci_last_digit_fast(n)
    if res_naive != res_test:
        print(n, res_naive, res_test)
        flag = False
    else:
        flag = True
```

```python
# 03. Greatest Common Divisor

def gcd_naive(a, b):
    current_gcd = 1
    for d in range(2, min(a, b) + 1):
        if a % d == 0 and b % d == 0:
            if d > current_gcd:
                current_gcd = d

    return current_gcd


def gcd_fast(a, b):
    n_max = max(a, b)
    n_min = min(a, b)
    if n_min == 0 and n_max != 0:
        return n_max
    elif n_min == 0 and n_max == 0:
        return 1
    elif n_min == n_max:
        return n_max
    elif n_max % n_min == 0:
        return n_min
    elif (n_max - n_min) < (n_max // n_min):
        return gcd_naive(n_max, n_min)
    else:
        while n_min != n_max:
            reminder = n_max - n_min
            if reminder == n_min:
                return reminder
            else:
                if reminder > n_min:
                    n_max = reminder
                else:
                    n_max = n_min
                    n_min = reminder
            if (n_max // n_min) > n_min:
                return gcd_naive(n_max, n_min)
        return reminder
    
# test
flag = True
while flag:
    a = random.randint(1, 10000)
    b = random.randint(1, 10000)
    res_naive = gcd_naive(a, b)
    res_test = gcd_fast(a, b)
    if res_naive != res_test:
        print(a, b, res_naive, res_test)
        flag = False
    else:
        flag = True
```

```python
# 04. Least Common Multiple

def lcm_naive(a, b):
    for l in range(1, a*b + 1):
        if l % a == 0 and l % b == 0:
            return l

    return a*b

def lcm_fast(a, b):
    if a < b:
        n_min = a
        n_max = b
    elif a > b:
        n_max = a
        n_min = b
    else:
        return a
    if n_max % n_min == 0:
        return n_max
    else:
        for i in range(1, n_min + 1):
            l = n_max * i
            if l % n_min == 0:
                return l

# test
flag = True
while flag:
    a = random.randint(1, 10**2)
    b = random.randint(1, 10**2)
    res_naive = lcm_naive(a, b)
    ###
    start_time = time.time()
    res_test = lcm_fast(a, b)
    delta = time.time() - start_time
    print("--- %s seconds ---" % delta)
    ###
    if res_naive != res_test:
        print(a, b, res_naive, res_test)
        flag = False
    else:
        flag = True
```
