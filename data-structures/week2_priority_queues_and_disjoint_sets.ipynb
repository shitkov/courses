```python
# 3.1 Make Heap

def sift_down(data, size, swaps, i):
    min_index = i
    left = 2 * i + 1
    right = 2 * i + 2
    if left < size and data[left] < data[min_index]:
        min_index = left
    if right < size and data[right] < data[min_index]:
        min_index = right
    if min_index != i:
        swaps.append((i, min_index))
        data[i], data[min_index] = data[min_index], data[i]
        sift_down(data, size, swaps, min_index)


def build_heap(data, swaps):
    size = len(data)
    for i in range(size // 2 - 1, -1, -1):
        sift_down(data, size, swaps, i)


import os
path = './week2_priority_queues_and_disjoint_sets/1_make_heap/tests/'
tests = os.listdir(path)
tests.sort()

for i, test in enumerate(tests):
    with open(path + test, 'r') as f:
        n = int(f.readline())
        data = list(map(int, f.readline().split()))
        assert len(data) == n

        swaps = []
        build_heap(data, swaps)

        print(len(swaps))
        for i, j in swaps:
            print(i, j)
        
    break
```
